De este diagnostico se desprenden las siguientes actividades:

- Cambiar los nombres de columnas a 'snakecase'
- Columna name:
    - Estandarizarla de forma que todos los nomnres en mayusculas , no existan espacios no deseados y se quiten los dignos de pausa
    - Imputar los casos perdidos por 'NOT DEFINED'
- Columna platform:
    - Agrupar las opciones existentes en la tipologias de forma que se cree la columna platform_group
- Columan year_of_release: 
    -  Quitar calores perdidos dado que son pocos
    - Tranfromar los tipos de datos a entero
- Columna genre:
    - Importar los casos perdidos por 'NOT DEFINED'
- Columna user_score:
    - Sustituir 'tbd' por un valor perdido numerico
    - Transformar a tipo numero float
    - Multiplicar por 10 los valores de forma que se estandaricen las magnitudes con el otro score de la tabla
- Columna rating:
    - Cambiar las opciones existentes por otros valores que den una interpretacion mas semantica de los datos
- Generar una nueva columna total_sales
- Revisar y eliminar duplicados


```python
#Cambiar columnas a snake_case

df_games.columns = [x.lower().replace (' ', '_') for x in df_games.columns]

df_games.info()
```

Columnas names

```python
#mayusculas
df_games['name'] = df_games['name'].str.upper()

```

```python
#Espacios no deseados

df_games['name'] = df_games['name'].str.strip()
```

```python
#Quitar signos de pausa

df_games['name'] = df_games['name'].str.replace('.','').str.replace(',','').str.replace(':','').str.replace(';','')
```

```python
#identificar valores perdidos
df_games[df_games['name'].isna()]
```

```python
#Remplazar pro 'NOT DEFINED'
df_games['name'] =df_games['name'].fillna('NOT DEFINED')
```

Columna platform

```python
#Crear una funcion aplicar en esta columna

def GroupPlat (x):
    if x in ['3DS','DS','GB','GBA','GC','N64','NES','SNES','Wii','WiiU']:
        return 'Nintendo'
    elif x in ['PC','PCFX']:
        return 'PC'
    elif x in ['PS','PS2','PS3','PS4','PSP','PSV']:
        return 'PlayStation'
    elif x in ['X360','XB','XOne']:
        return 'Xbox'
    else:
        return 'Other'
```

```python
df_games['platform_group'] = df_games ['platform'].apply(GroupPlat)

df_games.sample(10)
```

```python
df_games['platform_group'].value_counts(dropna=False)
```

Columna year_of_release

```python
#Identificar los valores perdidos
df_games[df_games['year_of_release'].isna()]
```

```python
#Eliminar valores perdidos
df_games = df_games.dropna(subset='year_of_release')
df_games.info()
```

```python
df_games['year_of_release'].describe()
```

```python
#Tranforma la columna a tipo entero

df_games['year_of_release'] = df_games['year_of_release'].astype(int)
df_games.info()
```

Columna genre

```python
#Identificar valores perdidos

df_games[df_games['genre'].isna()]
```

```python
#Inputar los valores perdidos por 'NOT DEFINED'
df_games['genre'] =df_games['genre'].fillna('Not defined')
df_games.info()
```

Columna user_score

```python
#Cambiar los casos que contengas tbd

df_games.loc[df_games['user_score']== 'tbd', 'user_score']=np.nan

df_games['user_score'].unique()
```

```python
#Cambiar el tipo de variable a float 
df_games['user_score'] =df_games['user_score'].astype(float)

df_games.info()
```

```python
#Escalar la variable para hacerla comparable con el otro score
df_games[['user_score','critic_score']].describe()
```

```python
df_games['user_score'] = df_games['user_score']*10
df_games[['critic_score','user_score']].describe()
```

Columna rating

```python
#funcion de transformacion de letras a ratings semanticos

def rat_sem (x):
    if x == 'E' or x == 'K-A':
        return 'Everyone'
    elif x  == 'T':
        return 'Teens'
    elif x  == 'M':
        return 'Mature'
    elif x == 'E10+':
        return 'Everyone 10y+'
    elif x == 'EC':
        return 'Early Childhood'
    elif x == 'RP':
        return 'Pending'
    elif x == 'AO':
        return 'Adults Only'
    else:
        return 'Pending'
```

```python
df_games ['Rating_sem'] = df_games['rating'].apply(rat_sem)

df_games.sample(10)
```

Crear columna total_sales

```python
df_games['total_sales'] = df_games[['na_sales','eu_sales','jp_sales','other_sales']].sum(axis=1)

df_games.sample(10)
```

Tratar duplicados

```python
#duplicado expresos

df_games[df_games.duplicated()]
```

```python
#duplicados implicitos o a entidad
df_games[df_games[['name', 'platform','year_of_release']].duplicated()]
```

```python
df_games.query('name == "MADDEN NFL 13" and platform == "PS3"')
```

```python
#Eliminar duplicados

df_games = df_games.drop_duplicates(subset =['name', 'platform','year_of_release']).reset_index(drop=True)
```
